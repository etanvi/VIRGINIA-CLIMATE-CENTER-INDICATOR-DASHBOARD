```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import math
import seaborn as sns
```

```python
# Read the data into a DataFrame with tab delimiter
data = pd.read_csv("C:/Users/namra/Downloads/sorted_data.csv")

# Print the column names to verify
print(data.columns)
```

```python
data
```

```python
# Replace -999 and 0 with NaN in the entire DataFrame
data = data.replace([-999], np.nan)
```

```python
# Deleting column cty
data.drop('cty', axis=1, inplace=True)
```

```python
data
```

```python
# Changing county column format
data['county'] = data['county'].str.replace('VA: ', '', regex=False)
```

```python
data
```

```python
# Adding more columns for convinience 
data['date'] = pd.to_datetime(data['date'])

# Create new columns for year, month, and day
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
```

```python
data
```

```python
# Download gzip file locally to upload into Bigquery
data.to_csv('C:/Users/namra/Downloads/cleaned_data.csv.gz', index=False, compression='gzip')
```
